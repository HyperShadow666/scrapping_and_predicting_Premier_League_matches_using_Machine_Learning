### cleaning predictors

```python
# converting home- away data into numerical ones 
matches["venue_code"] = matches["venue"].astype("category").cat.codes
```

```python
matches
```

```python
# converting opponent data into numerical ones 
matches["opponent_code"] = matches["opponent"].astype("category").cat.codes
```

```python
matches
```

```python
matches["hour"] = matches["time"].str.replace(":+", "", regex=True).astype("int")
```

```python
matches
```

```python
matches["day_code"] = matches["date"].dt.dayofweek
```

```python
matches
```

```python
matches["target"] = (matches["result"] == "W").astype("int")
```

```python
#using random forest since we are dealing with none linear data
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=70, min_samples_split= 10, random_state=1)
```

```python
train = matches[matches["date"] < '2023-01-01']
```

```python
test = matches[matches["date"] > '2023-01-01']
```

```python
predictors = ["venue_code", "opponent_code", "hour", "day_code"]
```

```python
rf.fit(train[predictors], train["target"])
```

```python
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(test["target"], preds)
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3,closed = 'left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset = new_cols)
    return group
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "fk","pk","pkatt"]
new_cols = [f"{c}_rolling" for c in cols]
```

```python
new_cols
```

```python
rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel('team')
```

```python
matches_rolling
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
matches_rolling
```

```python
def make_prediction(data, predictors):
    train = data[data["date"] < '2023-01-01']
    test =  data[data["date"] > '2023-01-01']
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], predicted = preds))
    precision = precision_score(test["target"], preds)
    return combined, precision
```

```python
combined, precision = make_prediction(matches_rolling, predictors + new_cols)
```
